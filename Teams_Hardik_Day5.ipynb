```python
import pandas as pd
import numpy as np
data=pd.read_csv('./match_data.csv')
data.head(10)
```

```python
no_of_match_as_team1=data.groupby('team1').size().reset_index(name="no_of_match_as_team1")
no_of_match_as_team1=no_of_match_as_team1.rename(columns={"team1":"team_name"})
no_of_match_as_team1
```

```python
no_of_match_as_team2=data.groupby('team2').size().reset_index(name="no_of_match_as_team2")
no_of_match_as_team2=no_of_match_as_team2.rename(columns={"team2":"team_name"})
no_of_match_as_team2
```

```python
no_of_matches=pd.merge(no_of_match_as_team1,no_of_match_as_team2,on="team_name")
no_of_matches
```

```python
no_of_matches['total_no_of_matches_played']=no_of_matches['no_of_match_as_team1']+no_of_matches['no_of_match_as_team2']
no_of_matches
```

```python
no_of_wins=data.groupby('winner').size().reset_index(name="no_of_wins")
no_of_wins=no_of_wins.rename(columns={'winner':'team_name'})
no_of_wins
```

```python
teams_df=pd.merge(no_of_matches,no_of_wins,on='team_name')
teams_df
```

```python
teams_df['%win']=(teams_df['no_of_wins']/teams_df['total_no_of_matches_played'])*100
teams_df
```

```python
matches_tie_data=data[data['result']=='tie']
matches_tie_data
```

```python
win_by_defending=data[data['result']=='runs'].groupby('winner').size().reset_index(name="wins_defend")
win_by_defending=win_by_defending.rename(columns={'winner':'team_name'})
win_by_defending
```

```python
win_by_chase=data[data['result']=='wickets'].groupby('winner').size().reset_index(name="wins_chase")
win_by_chase=win_by_chase.rename(columns={'winner':'team_name'})
win_by_chase
```

```python
team_df=pd.merge(teams_df,win_by_chase,on="team_name")
team_df=pd.merge(team_df,win_by_defending,on="team_name")
team_df
```

```python
del team_df['no_of_match_as_team1']
team_df
```

```python
del team_df['no_of_match_as_team2']
team_df
```

```python
team_df['%win_chase']=(team_df['wins_chase']/team_df['no_of_wins'])*100
team_df
```

```python
team_df['%win_defend']=(team_df['wins_defend']/team_df['no_of_wins'])*100
team_df
```
